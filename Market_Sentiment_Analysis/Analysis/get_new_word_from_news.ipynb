```python
import string
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def list_to_string(words):
    article=""
    for i in range(len(words)-1):
        article=article+" "+str(words[i])
        #print(article)
    article=article+str(words[len(words)-1])
    return article
def kill_punc(s):
    for c in string.punctuation:
        s= s.replace(c,"")
    #print(s)
    return s

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
 
    return False

for i in range(len(data["news"])):
    data["news"][i]=str(data["news"][i])
    words=data["news"][i].split(" ")
    for j in range(len(words)):
        #print(words[j])
        if is_number(words[j]) or words[j]=="$" or words[j]=="&" or words[j]==" " or words[j]=="the" or words[j]=="johnson" or words[j]=="reuters":
            words[j]=""
    #print(words)
        words[j]=str(words[j])
        if words[j] in stop_words:
                #print(words[j])
                words[j]=""
        if is_number(words[j]):
            print("a")
    words=list_to_string(words)
    data["news"][i]=words
    data["news"][i]=kill_punc(data["news"][i])
print(data["news"][1])
```

```python
def delete_the_same(l1):
    #list
    l2 = {}.fromkeys(l1).keys()
    return l2

from textblob import TextBlob
import nltk
positive=[]
neutral=[]
negative=[]
nltk.download('punkt')
for i in range(len(data["news"])):
    #for i in range(5):
    list=data["news"][i].split(" ")
    for j in range(len(list)):
        list[j]=str(list[j])
        blot=TextBlob(list[j])
        sentiment=blot.sentiment.polarity
        if sentiment>0:
            positive.append(list[j])
        elif sentiment==0:
            neutral.append(list[j])
        else:
            negative.append(list[j])
positive=delete_the_same(positive)
neutral=delete_the_same(neutral)
negative=delete_the_same(negative)
print("positive",positive)
print("neutral",neutral)
print("negative",negative)
```

```python
import string
def key_to_list(wordlist):
    newlist=[]
    for i in wordlist:
        i.lstrip("“")
        i.lstrip("‘")
        i.rstrip("”")
        i.rstrip("’")
        newlist.append(i)
    return newlist

print(type(positive))
positive_list=key_to_list(positive)
positive_DF=pd.DataFrame(positive_list)
print(positive_DF)
neutral_list=key_to_list(neutral)
neutral_DF=pd.DataFrame(neutral_list)
negative_list=key_to_list(negative)
negative_Df=pd.DataFrame(negative_list)
```

```python
writer = pd.ExcelWriter('word_list_news.xlsx')
positive_DF.to_excel(writer,'positive')
neutral_DF.to_excel(writer,'neutral')
negative_Df.to_excel(writer,'negative')
writer.save()
```
