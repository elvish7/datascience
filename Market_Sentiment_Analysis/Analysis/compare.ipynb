```python
import matplotlib.pyplot as plt
#set plot type
plt.rcParams['figure.figsize'] = (15, 5)
font = {'family': 'Segoe Script', 'weight': 'normal', 'size': 15}
plt.rc('font',**font)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rcParams['axes.unicode_minus']=False #show minus
```

```python
#Medical & Health
price[price['Symbols'] == 'IHE']['close'].plot(label='IHE')
price[price['Symbols'] == 'IBB']['close'].plot(label='IBB')
price[price['Symbols'] == 'XLV']['close'].plot(label='XLV')
plt.legend()
```

```python
#Industry
fig=plt.figure()

price[price['Symbols'] == 'DIA']['close'].plot(label='DIA')
price[price['Symbols'] == 'XLI']['close'].plot(label='XLI',secondary_y=True)

handles,labels = [],[]
for ax in fig.axes:
    for h,l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)

plt.legend(handles,labels)
```

```python
#Technology
fig=plt.figure()

ax1 = price[price['Symbols'] == 'QQQ']['close'].plot(label='QQQ')
ax2 = price[price['Symbols'] == 'IYZ']['close'].plot(secondary_y=True,label='IYZ')
ax3 = price[price['Symbols'] == 'SMH']['close'].plot(label='SMH')

handles,labels = [],[]
for ax in fig.axes:
    for h,l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)

plt.legend(handles,labels)
```
