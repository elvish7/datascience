```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)
import datetime  # For datetime objects
import os.path  # To manage paths
import sys  # To find out the script name (in argv[0])
import pandas as pd
# Import the backtrader platform
import backtrader as bt
#plt.figure(figsize=(15,5))
```

```python
if __name__ == '__main__':
    
    cerebro = bt.Cerebro() #creat the "brain"
    cerebro.broker.setcash(100000.0) #the "manager"
    
    #add the data and feed to Cerebro
    start = datetime.datetime(2010, 1, 1)
    end = datetime.datetime(2016, 10, 31)
    data = bt.feeds.YahooFinanceData(dataname='DIA', fromdate=start, todate=end)

    cerebro.adddata(data)
    
    print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())

    cerebro.run() #no strategy yet
    
    print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())
    
    
    cerebro.plot() #plot the result
```
