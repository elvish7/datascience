```python
import jieba #使用繁中字典
import jieba.analyse
#jieba.set_dictionary('data/dict.txt.big') #always error and I don't know why
# get keywords
# topK:the number of picked key words 
with open("titles.txt", "rb") as f:
    for line in f:
        tags = jieba.analyse.extract_tags(line, topK=5, withWeight=True)
        for tag, weight in tags:
            print(tag + "," + str(int(weight * 10000)))
```

```python
text = open("titles.txt", encoding = 'UTF-8').read()

# 設定停用字(排除常用詞、無法代表特殊意義的字詞) in fact doesn't work 
stopwords = {}.fromkeys(["檔案","表格","現場","不要"])

#mask
money_icon = np.array(Image.open( "money.png"))

# creat wordcloud
wc = WordCloud(font_path="C:\Windows\Fonts\HanyiSentyChalk.ttf",  #設置字體
               background_color="white", #背景顏色
               max_words = 2000 ,   #文字雲顯示最大字詞
               mask=money_icon,        
               stopwords = stopwords )#停用字詞
              # contour_width = 3, 
              # contour_color='steelblue')   
        
wc.generate(text)
```

```python
image_colors = ImageColorGenerator(money_icon)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()
#plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```
