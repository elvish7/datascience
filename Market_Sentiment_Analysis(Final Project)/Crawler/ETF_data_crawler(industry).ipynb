```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import matplotlib.pyplot as plt
from pandas_datareader import data
import numpy as np

etf = ['DIA','SPY','QQQ','IWM','IYR','XHB','ITB','IYT','ITA','IYC','IYZ','IHE',
      'IBB','RTH','SMH','XLB','XLE','XLF','XLI','XLP','XLU','XLV','XLY']

# collect etf data from 2015-05-01 to 2019-04-28
start = '2015-05-01'
end = '2019-04-28'
ticker = data.DataReader(etf, 'iex', start=start, end=end)

ticker.info()
ticker.to_csv('industry_etf.csv',na_rep='NA')
```
