```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import matplotlib.pyplot as plt
from pandas_datareader import data
import numpy as np

etf = ['DIA','SPY','QQQ','IVV',
       'XLI','XLV','XLK','XLE','XLF', #spdr sector
       'IYW','IYF','IYE','IYH','IYJ', # ishares sector
       'IYC','IYZ','IHE','IBB','RTH','SMH','XLB','XLP','XLY']

# collect etf data from 2015-05-01 to 2019-04-28
start = '2014-06-01'
end = '2019-06-01'
ticker = data.DataReader(etf, 'iex', start=start, end=end)
ticker.to_csv('industry_etf.csv')
ticker.info()
```
